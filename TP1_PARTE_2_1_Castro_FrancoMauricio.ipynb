```python
%matplotlib inline  
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
cd vision_computadora_I/Material_TPs/TP1
```

```python
img1 = cv.imread('img1_tp.png', cv.IMREAD_GRAYSCALE)
img2 = cv.imread('img2_tp.png', cv.IMREAD_GRAYSCALE)

plt.imshow(img1, cmap='gray', vmin=0, vmax=255)
plt.show()

plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.show()
```

Se eligieron 25 binds, debido a que determina de manera adecuada la forma de la distribución de los valores.

```python
#histograma para la imagen 1
histograma, bins = np.histogram(img1, bins=25, range=(0, 255))
plt.figure()
plt.title("Histograma de escala de grises")
plt.xlabel("Valor")
plt.ylabel("Cantidad de píxeles")
plt.xlim([0.0, 256])
plt.plot(bins[0:-1], histograma)
```

```python
#histograma para la imagen 2
histograma, bins = np.histogram(img2, bins=25, range=(0, 255))
plt.figure()
plt.title("Histograma de escala de grises")
plt.xlabel("Valor")
plt.ylabel("Cantidad de píxeles")
plt.xlim([0.0, 256])
plt.plot(bins[0:-1], histograma)
```

En ambas imágenes, se observa: i)patrones similares de distribución y ii) las imágenes no están preprocesadas. 

En general, se observa una trimodalidad. La primera es en el rango de valores entre 25-50; la segunda entre 175-225 y la tercera en 255. Es decir, baja cantidad de sombras, cantidad considerable de tonos medio bajos y medio altos, y cantidad considerable de tonos altos (blancos).
